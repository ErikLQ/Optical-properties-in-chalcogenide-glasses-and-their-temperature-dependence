First we define a function that calculates the refractive index as a funcion of the parameters **b1**, **b2**, **b3**, **c1**, **c2**, and **c3**.

```python
def N_flir(wl,b1,b2,b3,c1,c2,c3): #wavelenght has to be in micron
   return (1+ b1*wl**2/(wl**2-c1) + b2*wl**2/(wl**2-c2)+ b3*wl**2/(wl**2-c3))**0.5
def Rn_mat(n1, n2, wl):
    return ((n1-n2)/(n1+n2))**2
def T_mat(n1,n2,wl):
    return n2/n1*(2*n1/(n1+n2))**2
```

Function that takes  the parameters **b1**, **b2**, **b3**, **c1**, **c2**, and **c3** as input, as well as the wavelength range **wl** as input and returns the reflectance

```python
def Rtot(wl,b1,b2,b3,c1,c2,c3): 
    
    N_mat = N_flir(wl,b1,b2,b3,c1,c2,c3)
    R1 = Rn_mat(1.0,N_mat,wl)
    R2 = Rn_mat(N_mat,1.0,wl)
    R1_prime = R2
    T1 = T_mat(1.0, N_mat, wl)
    return R1 + T1**2*R2/(1-R1_prime*R2)
```
